```python
import yfinance

STOCK_SYMBOL = "^GSPC"
START_DATE = "2013-01-01"
END_DATE = "2023-07-09"

dataframe = yfinance.download(tickers = STOCK_SYMBOL,
                              start = START_DATE,
                              end = END_DATE)

dataframe
```

```python
import numpy

DECREASED = 1

INCREASED = 0

dataframe["Volume Increase/Decrease"] = numpy.where(dataframe["Volume"].shift(-1) > dataframe["Volume"],
                                                    DECREASED,
                                                    INCREASED)

dataframe
```

```python
dataframe["Returns"] = dataframe["Adj Close"].pct_change()
dataframe = dataframe.dropna()

dataframe
```

```python
BUY  = 1

SELL = 0

dataframe["Buy/Sell"] = numpy.where(dataframe["Adj Close"].shift(-1) > dataframe["Adj Close"],
                                    BUY,
                                    SELL)

dataframe["High vs Low"] = (dataframe["High"] - dataframe["Low"])/dataframe["Low"]

dataframe
```

```python
dataframe["Open vs Close"] = (dataframe["Open"] - dataframe["Close"]) / dataframe["Open"]

dataframe
```

```python
X = numpy.array(dataframe.drop(["Buy/Sell", "Close"], axis = 1))

y = numpy.array(dataframe["Buy/Sell"])

X
```

```python
from sklearn.neural_network import BernoulliRBM

NUMBER_OF_COMPONENTS = 2

neural_network = BernoulliRBM(n_components = NUMBER_OF_COMPONENTS)

neural_network.fit(X)
```

```python
neural_network.score_samples(X)
```
